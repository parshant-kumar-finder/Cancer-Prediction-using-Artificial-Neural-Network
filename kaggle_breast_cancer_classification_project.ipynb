```python
#train test split
```

```python
X=df.drop("benign_0__mal_1",axis=1).values
y=df["benign_0__mal_1"].values
```

```python
# for scaling and normalization
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#creating model1 with no early stopping and no droputlayers
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train
```

```python
#model 
model1=Sequential()
model1.add(Dense(units=30,activation="relu"))
model1.add(Dense(units=15,activation="relu"))
model1.add(Dense(units=1,activation="sigmoid"))
model1.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
#training model1 with 500 epochs which will overfit the data
model1.fit(x=X_train,y=y_train,epochs=500,validation_data=(X_test,y_test),verbose=1)

```

```python
model1_loss=pd.DataFrame(model1.history.history)
```

```python
model1_loss.plot()
```

```python
#evaluate model1
```

```python
predictions1=model1.predict_classes(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions1))
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
```

```python
early_stop=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
#model2 with dropout layers and early stopping to prevent overfitting
model2=Sequential()
model2.add(Dense(units=30,activation="relu"))
model2.add(Dropout(0.5))
model2.add(Dense(units=15,activation="relu"))
model2.add(Dropout(0.5))
model2.add(Dense(units=1,activation="sigmoid"))
model2.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model2.fit(x=X_train,y=y_train,epochs=500,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop])
```

```python
model2_loss=pd.DataFrame(model2.history.history)
model2_loss.plot()
```

```python
predictions2=model2.predict_classes(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions2))
print("\n")
print(confusion_matrix(y_test,predictions2))
```
